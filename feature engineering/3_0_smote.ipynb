```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class= 1, weights=[0.98], random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns=['feature1','feature2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature1'],final_df['feature2'],c=final_df['target'])
```
