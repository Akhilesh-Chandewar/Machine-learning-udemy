#5 Numbers Summary and Box Plot

```python
# Minimum, Maximum, Meduian, Q1, Q3, IQR
```

```python
import numpy as np
```

```python
lst_marks = [78, 85, 92, 67, 74, 88, 91, 69, 73, 84, 95, 62, 77, 89, 90, 71, 68, 82, 96, 80,
             64, 93, 75, 70, 86, 79, 94, 66, 87, 83, 72, 97, 76, 65, 81, 98, 63, 99, 100, 60]
min, q1, med, q3, max = np.quantile(lst_marks, [0, 0.25, 0.5, 0.75, 1])
```

```python
print(f"Minimum: {min}")
print(f"Q1: {q1}")
print(f"Median: {med}")
print(f"Q3: {q3}")
print(f"Maximum: {max}")
```

```python
IQR = q3 - q1
print(f"IQR: {IQR}")
```

```python
lower_fence = q1 - 1.5 * IQR
upper_fence = q3 + 1.5 * IQR
print(f"Lower Fence: {lower_fence}")
print(f"Upper Fence: {upper_fence}")
```

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks)
```
